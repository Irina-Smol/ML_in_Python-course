**График scatterplot пытается показать зависимость между количеством дней на работе (days employed) и возрастом человека (DAYS_BIRTH) для всех, кроме безработных. Обратите внимание, что для этого графика Вам нужно будет сначала удалить безработных людей из набора данных. <img src="task_one.jpg">


```python
import warnings
warnings.simplefilter('ignore')

employed = df[df['DAYS_EMPLOYED']<0]  
print(df['DAYS_EMPLOYED'])
# Делаем значения положительными 
employed['DAYS_EMPLOYED'] = -1*employed['DAYS_EMPLOYED'] 
employed['DAYS_BIRTH'] = -1*employed['DAYS_BIRTH']
```

```python
sns.scatterplot(x='DAYS_BIRTH',y='DAYS_EMPLOYED',data=employed, linewidth=0,alpha=0.01)
```

### ЗАДАНИЕ: воссоздайте приведённый ниже график Categorical:

<img src='catplot_solution.png'>

**Этот график отображает информацию только для *половины всех людей*, имеющих доход в этом наборе данных. Причём мы берём "нижнюю" половину - тех, кто зарабатывает меньше. Графики boxplot отображены для каждого значения колонки NAME_FAMILY_STATUS, показывая распределения их общего дохода. Раскраска hue сделана по колонке "FLAG_OWN_REALTY". Замечание: чтобы взять только нижнюю половину распределения дохода, Вам нужно будет преобразовать dataframe или взять только его часть *перед* тем, как рисовать график.**

```python
# НАПИШИТЕ ВАШ КОД В ЭТОЙ ЯЧЕЙКЕ
bottom_half_income = df.nsmallest(n=int(0.5*len(df)),columns='AMT_INCOME_TOTAL') 
sns.boxplot(x='NAME_FAMILY_STATUS',y='AMT_INCOME_TOTAL',data=bottom_half_income,hue='FLAG_OWN_REALTY') 
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,title='FLAG_OWN_REALTY') 
plt.title('Income Totals per Family Status for Bottom Half of Earners')
```
